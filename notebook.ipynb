```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Polygon, Rectangle
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display, clear_output

import ipywidgets as widgets


plt.rcParams["font.family"] = "DejaVu Sans"

# =====================================================
# LOGIC OF THE PROBLEM
# =====================================================

from math import gcd
from collections import deque


def euclides_ext(a, b):
    old_r, r = a, b
    old_s, s = 1, 0
    old_t, t = 0, 1

    while r != 0:
        q = old_r // r

        old_r, r = r, old_r - q * r
        old_s, s = s, old_s - q * s
        old_t, t = t, old_t - q * t

    return old_r, old_s, old_t  # gcd, coef_a, coef_b


def es_soluble(n, m, k):
    return (k % gcd(n, m) == 0) and (0 <= k <= max(n, m))


def resolver(n, m, k):
    inicio = (0, 0)
    visitados = {inicio}
    cola = deque([(inicio, [(inicio, "Inicio")])])

    while cola:
        (x, y), camino = cola.popleft()

        if x == k or y == k:
            return camino

        t_ab = min(x, m - y)  # A -> B
        t_ba = min(y, n - x)  # B -> A

        sucesores = [
            ((n, y), "Llenar A"),
            ((x, m), "Llenar B"),
            ((0, y), "Vaciar A"),
            ((x, 0), "Vaciar B"),
            ((x - t_ab, y + t_ab), "Verter A->B"),
            ((x + t_ba, y - t_ba), "Verter B->A"),
        ]

        for nuevo, accion in sucesores:
            if nuevo not in visitados:
                visitados.add(nuevo)
                cola.append((nuevo, camino + [(nuevo, accion)]))

    return None


# =====================================================
# INTRODUCTION SCREEN
# =====================================================

title = widgets.HTML(
    """
    <h1 style="
    text-align:center;
    color:#1E3A8A;
    font-size:32px;">
    Problema de las jarras
    </h1>
    """
)


description = widgets.HTML( 
"""
<div style="
background:#F8FBFF;
padding:25px;
border-radius:15px;
font-size:17px;
">

<h2>¿Cuál es el problema?</h2>

El problema de las jarras consiste en encontrar una serie de movimientos
para medir una cantidad exacta de agua usando dos recipientes con diferentes
capacidades.

<br><br>

Las operaciones permitidas son:

<ul>
<li>Llenar completamente una jarra.</li>
<li>Vaciar completamente una jarra.</li>
<li>Transferir agua entre jarras hasta que una se llene o la otra quede vacía.</li>
</ul>

Selecciona las capacidades de las jarras, determina la capacidad que quieres calcular en alguna de las 2 jarras y en caso de que se pueda resolver, observa como el programa lo hace.

</div>
"""
)


cap1_input = widgets.IntText(
    value=4,
    description="Capacidad jarra 1:",
    style={"description_width":"initial"}
)

cap1_input.layout = widgets.Layout(
    margin="0 30px 0 0"
)


cap2_input = widgets.IntText(
    value=5,
    description="Capacidad jarra 2:",
    style={"description_width":"initial"}
)
cap2_input.layout = widgets.Layout(
    margin="0 30px 0 0"
)

cap3_input = widgets.IntText(
    value=5,
    description="Capacidad a encontrar:",
    style={"description_width":"initial"}
)


start_button = widgets.Button(
    description="▶ Iniciar simulación",
    button_style="primary",
    layout=widgets.Layout(
        width="220px",
        height="40px"
    )
)

NegativeInputMessage = widgets.HTML("""
<div  style="text-align: center;">
    <h2 style="color: #B91C1C;">
        <b>Alguna de las entradas es negativa, por favor ingrese valores positivos.</b>
    </h2>
</div>
""")

NegativeInputMessage.layout.display = "none"


output = widgets.Output()

# =====================================================
# CONCLUSION SCREEN
# =====================================================

titleC = widgets.HTML(
    """
    <h1 style="
    text-align:center;
    color:#1E3A8A;
    font-size:32px;">
    ¡No existe solución!
    </h1>
    """
)


descriptionC = widgets.HTML(
"""
<div style="
background:#F8FBFF;
padding:25px;
border-radius:15px;
font-size:17px;
">

<h2>¿Por qué no existe?</h2>

Al hablar del problema de las jarras, en realidad estamos resolviendo una ecuación diofántica lineal de la forma:

<br><br>

ax + by = c

<br><br>

Donde todos los coeficientes son enteros y c tiene que ser menor que a o b. Sin embargo, dada la identidad de Bézout, esta ecuación tiene solución únicamente si c es múltiplo del MCD(a, b). En el caso dado, alguna condición no se cumple, por lo que no existe una manera de obtener la cantidad de agua deseada utilizando las dos jarras.

<br><br>

Finalmente, este problema permite comprender la relación entre las ecuaciones diofánticas lineales y situaciones cotidianas, además de resaltar la importancia de entender las propiedades matemáticas que las sustentan.

</div>
"""
)

return_button = widgets.Button(
    description="▶ Volver al menu",
    button_style="primary",
    layout=widgets.Layout(
        width="220px",
        height="40px"
    )
)

screenC = widgets.VBox(
    [
        titleC,
        descriptionC,


        widgets.HBox(
            [return_button],
            layout=widgets.Layout(
                justify_content="center",
                margin="15px 0px"
            )
        ),

        output
    ],
    layout=widgets.Layout(
        width="1200px",
        padding="30px",
        margin="20px auto 10px auto",
        align_items="center",
        border="2px solid #C7E7FF"
    )
)

# =====================================================
# START BUTTON
# =====================================================

def start_animation(button):

    capacity1 = cap1_input.value
    capacity2 = cap2_input.value
    spectedCapacity = cap3_input.value

    if (capacity1 <= 0 or capacity2 <= 0 or spectedCapacity <= 0):
        NegativeInputMessage.layout.display = "block"
        return

    clear_output()

    if es_soluble(capacity1, capacity2, spectedCapacity):

        path = resolver(capacity1, capacity2, spectedCapacity)
        animation = create_animation(
            cap1_input.value,
            cap2_input.value,
            path
        )

        html = animation.to_jshtml()

        display(HTML(f"""
        <div style="display:flex; justify-content:center;">
            {html}
        </div>
        """))
    else:
        display(screenC)

# =====================================================
# RETURN BUTTON
# =====================================================

def return_menu(button):

    clear_output()
    display(screen)

def create_animation(CAP1, CAP2, steps):


    # ----------------------------------------
    # States
    # ----------------------------------------


    # ----------------------------------------
    # Frames
    # ----------------------------------------

    frames=[]
    messages=[]

    frames_per_transition=25
    pause_frames=15


    for i in range(len(steps)-1):

        (a1,a2),_=steps[i]
        (b1,b2),message=steps[i+1]


        for t in np.linspace(
            0,
            1,
            frames_per_transition,
            endpoint=False
        ):

            frames.append(
                (
                    a1+(b1-a1)*t,
                    a2+(b2-a2)*t
                )
            )

            messages.append(message)


        for _ in range(pause_frames):

            frames.append((b1,b2))
            messages.append(message)



    frames.append(steps[-1][0])
    messages.append(steps[-1][1])


    # ----------------------------------------
    # Figure
    # ----------------------------------------

    fig,ax=plt.subplots(
        figsize=(9,6)
    )


    fig.patch.set_facecolor("#EAF6FF")
    ax.set_facecolor("#EAF6FF")

    ax.set_xlim(0,10)
    ax.set_ylim(0,8)

    ax.axis("off")
    ax.set_aspect("equal")



    # ----------------------------------------
    # Jar shape
    # ----------------------------------------

    def create_jar(x,y,width,height):

        neck_width=width*0.55

        neck_height=height*0.18

        shoulder=height*0.20


        return Polygon(
        [
            (x,y),
            (x+width,y),

            (x+width+0.18,y+shoulder),
            (x+width+0.18,y+height-neck_height),

            (x+width*0.75,y+height-neck_height/2),
            (x+width*0.75,y+height),

            (x+width*0.25,y+height),
            (x+width*0.25,y+height-neck_height/2),

            (x-0.18,y+height-neck_height),
            (x-0.18,y+shoulder)
        ],

        closed=True,
        facecolor=(1,1,1,0.15),
        edgecolor="#555555",
        linewidth=3,
        zorder=3
        )


    max_cap=max(CAP1,CAP2)


    h1=5*CAP1/max_cap
    h2=5*CAP2/max_cap



    jar1=create_jar(
        2,1,1.5,h1
    )

    jar2=create_jar(
        6,1,1.5,h2
    )


    ax.add_patch(jar1)
    ax.add_patch(jar2)



    # ----------------------------------------
    # Water
    # ----------------------------------------

    water1=Rectangle(
        (1.8,1),
        1.9,
        0,
        facecolor="#2A9DF4",
        zorder=1
    )


    water2=Rectangle(
        (5.8,1),
        1.9,
        0,
        facecolor="#2A9DF4",
        zorder=1
    )


    water1.set_clip_path(jar1)
    water2.set_clip_path(jar2)


    ax.add_patch(water1)
    ax.add_patch(water2)



    # ----------------------------------------
    # Text
    # ----------------------------------------

    ax.text(
        5,
        7.5,
        "Problema de las jarras",
        fontsize=26,
        fontweight="bold",
        color="#1E3A8A",
        ha="center",
        zorder=10
    )


    action=ax.text(
        5,
        6.8,
        "",
        fontsize=16,
        color="#2563EB",
        ha="center",
        style="italic"
    )


    label1=ax.text(
        2.75,
        0.4,
        "",
        ha="center",
        fontsize=14
    )


    label2=ax.text(
        6.75,
        0.4,
        "",
        ha="center",
        fontsize=14
    )

    label3=ax.text(
        2.75,
        -0.25,
        "A",
        ha="center",
        fontsize=14
    )


    label4=ax.text(
        6.75,
        -0.25,
        "B",
        ha="center",
        fontsize=14
    )



    # ----------------------------------------
    # Update
    # ----------------------------------------

    def update(frame):

        amount1,amount2=frames[frame]


        water1.set_height(
            h1*amount1/CAP1
        )


        water2.set_height(
            h2*amount2/CAP2
        )


        label1.set_text(
            f"{amount1:.0f}/{CAP1} L"
        )


        label2.set_text(
            f"{amount2:.0f}/{CAP2} L"
        )


        action.set_text(
            messages[frame]
        )


        return (
            water1,
            water2,
            label1,
            label2,
            action
        )



    ani=FuncAnimation(
        fig,
        update,
        frames=len(frames),
        interval=40,
        blit=True,
        repeat=False
    )
  
    plt.close(fig)

    return ani


# =====================================================
# CONNECT BUTTON
# =====================================================

start_button.on_click(start_animation)
return_button.on_click(return_menu)

display(widgets.HTML("""
<style>
.jp-OutputArea {
    background: linear-gradient(135deg, #DFF6FF, #EDF7FF);
    padding: 25px;
    border-radius: 20px;
}
</style>
"""))

screen = widgets.VBox(
    [
        title,
        description,

        widgets.HBox(
            [cap1_input, cap2_input, cap3_input],
            layout=widgets.Layout(
                justify_content="center",
                margin="20px 0px"
            )
        ),

        widgets.HBox(
            [start_button],
            layout=widgets.Layout(
                justify_content="center",
                margin="15px 0px"
            )
        ),

        NegativeInputMessage,

        output
    ],
    layout=widgets.Layout(
        width="1200px",
        padding="30px",
        margin="20px auto 10px auto",
        align_items="center",
        border="2px solid #C7E7FF"
    )
)

screen.add_class("my-container")

display(HTML("""
<style>
.my-container{
    background:#DCEFFD;
    border-radius:20px;
    padding:30px;
}
</style>
"""))

display(screen)
```
